```python
import numpy as np
from pylfsr import LFSR
```

$f(x) = x^{7} + x +1$

```python
seed = [1, 0, 0, 1, 0, 1, 0]
fpoly = [7, 1]
```

```python
L = LFSR(fpoly=fpoly,initstate =seed)
seq = L.runKCycle(10)
```

```python
print('count \t state \t\toutbit \t seq')
print('-'*50)
for _ in range(15):
    print(L.count,L.state,'',L.outbit,L.seq,sep='\t')
    L.next()
print('-'*50)
print('Output: ',L.seq)
```

```python
L.Viz(show=False, show_labels=False,title='R1')
```
